## 1- Polynomial regression / overfitting / regularization

### Load the data "ex2data1.csv"

```python
data1 = pd.read_csv('./ex2data1.csv')
```

```python
data1.head(5)
```

```python
data1 = data1.sort_values(by=['x'])
```

```python
X = data1["x"]
Y = data1["y"]
```

### Visualize the data

```python
plt.scatter(X,Y,color="red",marker="x")
plt.title("Scatter Plot")
plt.show()
```

### Convert the dataframe into matrix

```python
data1_array = np.array(data1)
```

```python
X_mat = data1_array[::, :1]
Y_mat = data1_array[::, 1:]
```

### Normalize the data

```python
X_mat_norm = stats.zscore(X_mat)
```

```python
bias = np.ones((X_mat_norm.shape[0],X_mat_norm.shape[1]+1))
bias[:,1:] = X_mat_norm
X_b = bias
np.random.seed(1)
weights_first = np.random.rand(2,1)
```

### Cost function

```python
def computeCost(X,Y,weights):
    m = X.shape[0]
    n = X.shape[1]
    prediction = np.dot(X, weights)
    return (1/m)*((np.square(prediction-Y)).sum(axis=0))
```

### Gradient Descent Implementation 

```python
def gradientDescent(X, Y, weights, alpha, iterations):
    m = len(Y)  
    count = 1
    cost_history = []
    while(count <= iterations):
        prediction = np.dot(X, weights)
        cost = prediction - Y
        grad = (2/m)*(np.dot(X.transpose(), cost))
        weights = weights - (alpha*grad)
        cost_history.append(computeCost(X, Y, weights)[0])
        count = count + 1
    min_cost_iter = cost_history.index(min(cost_history)) + 1
    return weights, cost_history, min(cost_history)
```

## 1-1 First order Regression Model
y = W0 + W1 * x

#### fitting data using learning rate 0.01

```python
weights_first, cost_history_first, min_cost_first = gradientDescent(X_b, Y_mat, weights_first, 0.01, 2000)
print ("Minimum cost achieved with learning rate 0.01 is ", min_cost_first)
```

### Optimized Weights

```python
print ("The final weights after optimization are:")
print ("W0 = ", weights_first[0][0])
print ("W1 = ", weights_first[1][0])
```

### Plot the data with the fitted line

```python
prediction_first = np.dot(X_b, weights_first)
```

```python
plt.scatter(X,Y,40,color="red",marker="x")
plt.plot(X,prediction_first)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression best line fitted to training examples")
plt.show()
```

### Root Mean Square Error

```python
RMSE_first = math.sqrt((1/len(Y_mat))*((np.square(prediction_first-Y_mat)).sum(axis=0)))
print ("Root Mean Square Error using First Order Regression Model: ", RMSE_first)
```

## 1-5 Sixteenth order Regression Model With Regularization

### Cost Function with Regularization

```python
def computeCostReg(X,Y,weights,lamdha):
    m = X.shape[0]
    n = X.shape[1]
    prediction = np.dot(X, weights)
    J = (1/m)*((np.square(prediction-Y)).sum(axis=0))
    weights_r = weights[1:]
    J = J + (lamdha/m)* weights_r.sum()
    return J
```

### Gradient Descent with Regularization

```python
def gradientDescentReg(X, Y, weights, alpha, iterations, lamdha):
    m = len(Y)  
    count = 1
    cost_history = []
    while(count <= iterations):
        prediction = np.dot(X, weights)
        cost = prediction - Y
        grad = (2/m)*(np.dot(X.transpose(), cost))
        weights_r = weights[1:]
        grad_r = grad[1:]
        grad_r = grad_r + 2*(lamdha/m)* weights_r
        grad[1:] = grad_r
        weights = weights - (alpha*grad)
        cost_history.append(computeCostReg(X, Y, weights, lamdha)[0])
        count = count + 1
    min_cost_iter = cost_history.index(min(cost_history)) + 1

    return weights, cost_history, min(cost_history)
```

### weight optimization using learning rate 0.01 and lambda 0.1

```python
weights_reg_01, cost_history_reg_01, min_cost_reg_01 = gradientDescentReg(X_b_16, Y_mat_16, weights_16, 0.01, 1000, \
                                                                          0.1)
print ("Minimum cost achieved with learning rate 0.01 and lambda 0.1 is ", min_cost_reg_01)
```

### Function to generate Ridge Coefficients as a function of lambda

```python
def generate_ridge(X, Y, lamdha = [0.1,1,5,10,20,30,40,50,60,70]):
    weights = []
    for l in lamdha: 
        regr = Ridge(alpha=l, max_iter=1000)
        regr.fit(X, Y) 
        weights.append(list(regr.coef_[0]))

    return lamdha, list(zip(*weights))
```

### Function to get Lasso Coefficients as a function of lambda

```python
def generate_lasso(X, Y,lamdha=[0.0005,0.001,0.01,0.1,1,10,20,30, 40,50, 60, 70, 80, 1000, 10000, 20000, 80000]):
    weights = [] 
    for l in lamdha: 
        regr = Lasso(alpha=l, max_iter=10000)
        regr.fit(X, Y) 
        weights.append(list(regr.coef_))

    return lamdha, list(zip(*weights))
```

### Plotting Lasso Coefficients as a function of lambda

```python
l_16, weights_16 = generate_lasso(X2_train_16[:, 1:], Y2_train_16)

colors = mpl.cm.rainbow(np.linspace(0, 1, 16))
fig, ax = plt.subplots()
for color, w in zip(colors, weights_16[1:]):
    ax.plot(l_16, w, color=color)
plt.xlabel("lambda")
plt.ylabel("weights")
plt.title("Weights as a function of Regularization using Lasso")
plt.show()
```

## 3-3. Fit the training data using regression model with ridge (L2 penalty) regularization using TensorFlow

```python
X = tf.placeholder(tf.float32,[None,X_train_tf.shape[1]])
Y = tf.placeholder(tf.float32,[None,1])
```

### L2 Regularization 

```python
W_tf = tf.Variable(tf.random_uniform([n+1,1], -1.0, 1.0))
prediction_train = tf.matmul(X, W_tf)
error_train = prediction_train - Y
rmse_tf_train = tf.sqrt(tf.reduce_mean(tf.square(error_train)))
cost = tf.reduce_mean(tf.square(error_train))
reg_factor = tf.reduce_sum(tf.square(W_tf))
lamdha = 0.1
alpha = 0.01
cost_reg = tf.add(cost, (lamdha/m)*reg_factor)
gradients = (2/m)*tf.add(tf.matmul(tf.transpose(X), error_train), (lamdha*W_tf))
training_operation = tf.assign(W_tf, W_tf - alpha*gradients)
epochs =30000
```

### using lambda 0.1

```python
cost_history_tf_01 = []
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(epochs):
        sess.run(training_operation, feed_dict={X:X_train_tf, Y:Y_train_tf})
        cost_history_tf_01.append(sess.run(cost_reg, feed_dict={X:X_train_tf, Y:Y_train_tf}))
        
    pred_train_01 = sess.run(prediction_train, feed_dict={X: X_train_tf})
    rmse_train_01 = math.sqrt(sess.run(rmse_tf_train, feed_dict={X:X_train_tf, Y:Y_train_tf}))
    pred_valid_01 = sess.run(prediction_train, feed_dict={X: X_valid_tf})
    rmse_valid_01 = math.sqrt(sess.run(rmse_tf_train, feed_dict={X:X_valid_tf, Y:Y_valid_tf}))
    print ("Root Mean Square Error on training dataset using lambda 0.1: ", rmse_train_01)
    print ("Root Mean Square Error on validation dataset using lambda 0.1: ", rmse_valid_01)
```
